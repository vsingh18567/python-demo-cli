```python

with open("benchmark.txt") as f:
    lines = f.readlines()

curr_size = None

sizes = []
times = {}
for line in lines:
    if "generate" in line:
        parts = line.split(" ")
        parts = [part for part in parts if part != ""]
        func_name = parts[3]
        mean = float(parts[5])
        if func_name not in times:
            times[func_name] = []
        times[func_name].append(mean)

    if "Results" in line:
        curr_size = int(line.split(" ")[-1])
        sizes.append(curr_size)
```
